Copias instantneas, no diferencias

La principal diferencia entre Git y cualquier otro VCS (Sistema de control de versiones) es la forma en la que manejan sus datos. 
Git maneja sus datos como un conjunto de copias instantneas de un sistema de archivos miniatura. Cada vez que se confirma un cambio, o se guarda el estado de un proyecto en Git, bsicamente toma una foto del aspecto de todos los archivos en ese momento y guarda una referencia a esa copia instantnea. 
Para ser eficiente, si los archivos no se han modificado Git no almacena el archivo de nuevo, utiliza un enlace al archivo anterior idntico que ya tiene almacenado. Git maneja sus datos como una secuencia de copias instantneas.
